```python
# Core libraries
import pandas as pd
import numpy as np

# NLP & ML
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Text cleaning
import re
```

```python
def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text
```

```python
job_description = """
Looking for a Data Analyst skilled in Python, SQL, Power BI, and Machine Learning.
"""

resumes = [
    "Experienced in Python, SQL, and data visualization.",
    "Worked with Java and web development.",
    "Strong background in Python, Machine Learning, and data analysis."
]
```

```python
job_description_clean = clean_text(job_description)
resumes_clean = [clean_text(resume) for resume in resumes]
```

```python
# Combine JD + resumes
all_texts = [job_description_clean] + resumes_clean

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(all_texts)
```

```python
jd_vector = vectors[0]

scores = []

for i in range(1, vectors.shape[0]):
    score = cosine_similarity(jd_vector, vectors[i])[0][0]
    scores.append(score)
```

```python
results = pd.DataFrame({
    "Resume": resumes,
    "Score": scores
})

results["Score"] = results["Score"] * 100  # convert to %
results = results.sort_values(by="Score", ascending=False)
results["Score"] = results["Score"].round(2)
results.reset_index(drop=True, inplace=True)

results

results
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(file_path):
    text = ""
    doc = fitz.open(file_path)
    
    for page in doc:
        text += page.get_text()
    
    return text
```

```python
import os

folder_path = "resume"

resumes = []

for file in os.listdir(folder_path):
    if file.endswith(".pdf"):
        full_path = os.path.join(folder_path, file)
        text = extract_text_from_pdf(full_path)
        resumes.append(text)
```

```python
resumes_clean = [clean_text(resume) for resume in resumes]
job_description_clean = clean_text(job_description)
```

```python
all_texts = [job_description_clean] + resumes_clean

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(all_texts)

jd_vector = vectors[0]

scores = []

for i in range(1, vectors.shape[0]):
    score = cosine_similarity(jd_vector, vectors[i])[0][0]
    scores.append(score)
```

```python
resume_files = []

for file in os.listdir(folder_path):
    if file.endswith(".pdf"):
        full_path = os.path.join(folder_path, file)
        text = extract_text_from_pdf(full_path)
        resumes.append(text)
        resume_files.append(file)
```

```python
results = pd.DataFrame({
    "Resume": resume_files,
    "Score": scores
})

results["Score"] = (results["Score"] * 100).round(2)
results = results.sort_values(by="Score", ascending=False).reset_index(drop=True)

results
```

```python
skills_list = [
    "python", "sql", "machine learning", "data analysis",
    "power bi", "excel", "tableau", "statistics",
    "deep learning", "nlp"
]
```

```python
def extract_skills(text, skills_list):
    found_skills = []
    
    for skill in skills_list:
        if skill in text:
            found_skills.append(skill)
    
    return found_skills
```

```python
jd_skills = extract_skills(job_description_clean, skills_list)
print("JD Skills:", jd_skills)
```

```python
resume_skills = []

for resume in resumes_clean:
    skills = extract_skills(resume, skills_list)
    resume_skills.append(skills)
```

```python
matched_skills = []
missing_skills = []

for skills in resume_skills:
    matched = list(set(skills) & set(jd_skills))
    missing = list(set(jd_skills) - set(skills))
    
    matched_skills.append(matched)
    missing_skills.append(missing)
```

```python
results["Matched Skills"] = matched_skills
results["Missing Skills"] = missing_skills

results
```
